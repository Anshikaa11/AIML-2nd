```python
import pandas as pd
data = {
    "Question": [
        "What is the capital of India?",
        "Which language is used for Data Science?",
        "What is 2 + 2?",
        "Who is the father of Python?"
    ],
    "Answer": [
        "Delhi",
        "Python",
        "4",
        "Guido van Rossum"
    ]
}
quiz = pd.DataFrame(data)
score = 0
print("---- Welcome to Simple Quiz ----\n")
for i in range(len(quiz)):
    user_ans = input(quiz.loc[i, "Question"] + " : ")
    if user_ans.strip().lower() == quiz.loc[i, "Answer"].lower():
        print("Correct\n")
        score += 1
    else:
        print("Wrong")
        print("Correct Answer:", quiz.loc[i, "Answer"], "\n")

print("Quiz Finished")
print("Your Score:", score, "/", len(quiz))
```
