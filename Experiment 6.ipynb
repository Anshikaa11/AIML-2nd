```python
import pandas as pd
import numpy as np
data = pd.read_excel(r"C:\Users\Anshika\Downloads\datafinal.csv.xlsx")
print("Number of Rows:", data.shape[0])
print("Number of Columns:", data.shape[1])
print("Dataset Size:", data.size)
print("\nFirst Five Rows:")
print(data.head())
print("\nMissing Values in Each Column:")
print(data.isnull().sum())
num_cols = ['Age', 'Salary', 'Experience', 'Rating']
print("\nSum of Numerical Columns:")
print(data[num_cols].sum())
print("\nAverage (Mean) of Numerical Columns:")
print(data[num_cols].mean())
print("\nMinimum Values:")
print(data[num_cols].min())
print("\nMaximum Values:")
print(data[num_cols].max())
data[num_cols] = data[num_cols].fillna(data[num_cols].mean())
for col in ['Gender', 'City', 'State']:
    data[col].fillna(data[col].mode()[0], inplace=True)
print("\nMissing values after handling:")
print(data.isnull().sum())
for col in num_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    data[col] = np.clip(data[col], lower, upper)
data.to_csv("cleaned_lab4_dataset.csv", index=False)
print("\nData preprocessing completed successfully.")
```
