```python
# Step 1: Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Set style for seaborn
sns.set_style("whitegrid")

# Step 3: Load Dataset
file_path = r"C:\Users\Anshika\.cache\kagglehub\datasets\mlg-ulb\creditcardfraud\versions\3\creditcard.csv"
data = pd.read_csv(file_path)

# Step 4: Display first 5 rows
print("First 5 rows of dataset:")
print(data.head())

# Step 5: Dataset information
print("\nDataset Info:")
print(data.info())

# Step 6: Statistical summary
print("\nStatistical Summary:")
print(data.describe())

# Step 7: Check missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Step 8: Class distribution (Fraud vs Non-Fraud)
plt.figure()
sns.countplot(x="Class", data=data)
plt.title("Fraud vs Non-Fraud Transactions")
plt.show()

# Step 9: Distribution of transaction amount
plt.figure()
sns.histplot(data["Amount"], bins=50, kde=True)
plt.title("Transaction Amount Distribution")
plt.show()

# Step 10: Distribution of transaction time
plt.figure()
sns.histplot(data["Time"], bins=50, kde=True)
plt.title("Transaction Time Distribution")
plt.show()

# Step 11: Box plot for amount vs class
plt.figure()
sns.boxplot(x="Class", y="Amount", data=data)
plt.title("Amount vs Fraud Class")
plt.show()

# Step 12: Correlation heatmap (for selected features)
plt.figure(figsize=(12,8))
corr = data.corr()
sns.heatmap(corr, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Step 13: Scatter plot for Amount vs Time
plt.figure()
sns.scatterplot(x="Time", y="Amount", hue="Class", data=data)
plt.title("Time vs Amount Scatter Plot")
plt.show()
```
