```python
import pandas as pd
import matplotlib.pyplot as plt

# 🔹 Load the CSV created earlier
df = pd.read_csv("C:/Users/Anshika/Downloads/ml-latest-small/ml-latest-small/movies.csv")

# =========================
# 🔍 BASIC INFO
# =========================
print("\n🔸 First 5 Rows:")
print(df.head())

print("\n🔸 Shape of Dataset:")
print(df.shape)

print("\n🔸 Columns:")
print(df.columns)

print("\n🔸 Data Types:")
print(df.dtypes)

print("\n🔸 Info:")
df.info()

# =========================
# ❗ MISSING VALUES
# =========================
print("\n🔸 Missing Values:")
print(df.isnull().sum())

# =========================
# 📈 STATISTICS
# =========================
print("\n🔸 Statistical Summary:")
print(df.describe())

# =========================
# 🔢 NUMERICAL ANALYSIS
# =========================
numeric_df = df.select_dtypes(include=['number'])

print("\n🔸 Mean:")
print(numeric_df.mean())

print("\n🔸 Median:")
print(numeric_df.median())

print("\n🔸 Min:")
print(numeric_df.min())

print("\n🔸 Max:")
print(numeric_df.max())

# =========================
# 🔗 CORRELATION
# =========================
print("\n🔸 Correlation Matrix:")
print(numeric_df.corr())

# =========================
# 📊 VISUALIZATION
# =========================

# Histogram
numeric_df.hist(figsize=(10, 8))
plt.suptitle("Histogram of Numerical Features")
plt.show()

# Boxplot
numeric_df.plot(kind='box', subplots=True, layout=(3,3), figsize=(10,8))
plt.suptitle("Boxplot (Outlier Detection)")
plt.show()

# Heatmap (basic)
plt.imshow(numeric_df.corr())
plt.colorbar()
plt.title("Correlation Heatmap")
plt.show()
```
