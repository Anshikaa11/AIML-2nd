```python
import pandas as pd
df = pd.read_excel(r"C:\Users\Anshika\Downloads\datafinal.csv.xlsx")
print("DATASET DETAILS")
print("-" * 40)
print(f"Number of rows    : {df.shape[0]}")
print(f"Number of columns : {df.shape[1]}")
print(f"Total size (cells): {df.size}")
print("\nFirst five rows:")
print(df.head())
print("\nMissing values in each column:")
print(df.isnull().sum())
numeric_df = df.select_dtypes(include=["number"])
print("\nNUMERICAL COLUMN STATISTICS")
print("-" * 40)
print("\nSum:")
print(numeric_df.sum())
print("\nAverage (Mean):")
print(numeric_df.mean())
print("\nMinimum values:")
print(numeric_df.min())
print("\nMaximum values:")
print(numeric_df.max())
output_file = "exported_data.csv"
df.to_csv(output_file, index=False)
print("\nData exported successfully to:", output_file)
```
