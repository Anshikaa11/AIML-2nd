```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 🔹 Load dataset
df = pd.read_csv('C:/Users/Anshika/.cache/kagglehub/datasets/mlg-ulb/creditcardfraud/versions/3/creditcard.csv')

print("🔸 Dataset Preview:")
print(df.head())

# =========================
# 🎨 MATPLOTLIB VISUALIZATIONS
# =========================

# 1. Line Plot
plt.figure()
plt.plot(df.select_dtypes(include='number').iloc[:, 0])
plt.title("Line Plot")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()

# 2. Bar Chart
plt.figure()
df.select_dtypes(include='number').iloc[:, 0].head(10).plot(kind='bar')
plt.title("Bar Chart")
plt.show()

# 3. Histogram
plt.figure()
plt.hist(df.select_dtypes(include='number').iloc[:, 0])
plt.title("Histogram")
plt.show()

# 4. Scatter Plot
num_cols = df.select_dtypes(include='number').columns
if len(num_cols) >= 2:
    plt.figure()
    plt.scatter(df[num_cols[0]], df[num_cols[1]])
    plt.title("Scatter Plot")
    plt.xlabel(num_cols[0])
    plt.ylabel(num_cols[1])
    plt.show()

# =========================
# 🌈 SEABORN VISUALIZATIONS
# =========================

# 5. Box Plot
plt.figure()
sns.boxplot(data=df.select_dtypes(include='number'))
plt.title("Box Plot")
plt.show()

# 6. Pair Plot
sns.pairplot(df.select_dtypes(include='number'))
plt.show()

# 7. Heatmap (Correlation)
plt.figure()
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()

# 8. Count Plot (for categorical data)
cat_cols = df.select_dtypes(include='object').columns
if len(cat_cols) > 0:
    plt.figure()
    sns.countplot(x=df[cat_cols[0]])
    plt.title("Count Plot")
    plt.show()

print("\n✅ Data Visualization Completed")
```
