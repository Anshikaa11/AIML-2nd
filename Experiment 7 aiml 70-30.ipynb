```python
# ==============================
# 1. Import Libraries
# ==============================
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# ==============================
# 2. Load Dataset
# ==============================
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df = pd.read_csv(url, sep=';')

# ==============================
# 3. Features & Target
# ==============================
X = df.drop('quality', axis=1)
y = df['quality']

# ==============================
# 4. 70-30 Split
# ==============================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# ==============================
# 5. Feature Scaling
# ==============================
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ==============================
# 6. Convert to DataFrame
# ==============================
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X.columns)

# ==============================
# 7. Plot Graph (Before vs After Scaling)
# ==============================

# Choose one feature (e.g., alcohol)
feature = 'alcohol'

plt.figure()

# Before scaling
plt.hist(X_train[feature], bins=20, alpha=0.5, label='Before Scaling')

# After scaling
plt.hist(X_train_scaled[feature], bins=20, alpha=0.5, label='After Scaling')

plt.xlabel(feature)
plt.ylabel("Frequency")
plt.title("Before vs After Feature Scaling (70-30 Split)")
plt.legend()

plt.show()
```
