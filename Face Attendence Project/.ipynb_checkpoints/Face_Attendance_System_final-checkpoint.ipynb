# Face Attendance System Dashboard

Welcome to the Python-based Interactive Face Attendance System!
This system uses your webcam to register users and track their attendance in real time.

**Core Features**:
- Register users via Webcam
- Live Face Tracking and automatic Attendance Logging
- Direct Pandas dataframe storage (`data/users.csv`, `data/attendance.csv`)

*Please run the cell below to load the Dashboard UI.*

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import pandas as pd
import matplotlib.pyplot as plt
from core.backend import FaceAttendanceSystem

# Initialize backend
system = FaceAttendanceSystem('data')

# Create Output area for logs and visuals
output_area = widgets.Output()

# --- Handlers ---

def on_register_click(b):
    with output_area:
        clear_output()
        name = name_input.value.strip()
        if not name:
            print("\u26a0\ufe0f Please enter a valid name.")
            return
        print(f"Starting registration for '{name}'... Check your taskbar for the webcam window.")
        success = system.register_user(name)
        if success:
            print("\u2705 User registered successfully!")
            name_input.value = ''
        else:
            print("\u274c Registration failed or was cancelled.")


def on_remove_click(b):
    with output_area:
        clear_output()
        name = remove_input.value.strip()
        if not name:
            print("⚠️ Please enter a valid name to remove.")
            return
        success = system.remove_user(name)
        if success:
            remove_input.value = ''
        else:
            print(f"❌ Could not remove user '{name}'.")

def on_track_click(b):
    with output_area:
        clear_output()
        print("\u23f1\ufe0f Starting live tracking... Check your taskbar for the webcam window.")
        system.run_tracking()
        print("\u23f9\ufe0f Tracking stopped.")

def on_view_users_click(b):
    with output_area:
        clear_output()
        system.load_data() # refresh data
        if system.users_df.empty:
            print("No users registered yet.")
        else:
            display(HTML("<h3>Registered Users</h3>"))
            display(system.users_df)

def on_view_attendance_click(b):
    with output_area:
        clear_output()
        system.load_data() # refresh data
        if system.attendance_df.empty:
            print("No attendance marked yet.")
        else:
            display(HTML("<h3>Attendance Logs</h3>"))
            # Show most recent first
            display(system.attendance_df.sort_values(by='timestamp', ascending=False).head(20))

def on_view_chart_click(b):
    with output_area:
        clear_output()
        system.load_data()
        if system.attendance_df.empty:
            print("Not enough data for a chart.")
            return
        
        # Count attendance per day
        counts = system.attendance_df.groupby('date').size()
        
        fig, ax = plt.subplots(figsize=(8, 4))
        counts.plot(kind='bar', color='cadetblue', ax=ax)
        ax.set_title("Attendance Count by Date", fontsize=14)
        ax.set_ylabel("Number of attendees")
        ax.set_xlabel("Date")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

def on_clear_logs_click(b):
    with output_area:
        clear_output()
        system.clear_attendance()

# --- UI Elements ---

style = {'description_width': 'initial'}

# Registration row
name_input = widgets.Text(placeholder='e.g. John Doe', description='Name:', style=style)
register_btn = widgets.Button(description='Capture & Register', button_style='primary', icon='camera')
register_btn.on_click(on_register_click)
register_box = widgets.HBox([name_input, register_btn])


# Remove User row
remove_input = widgets.Text(placeholder='e.g. John Doe', description='Name:', style=style)
remove_btn = widgets.Button(description='Remove User', button_style='danger', icon='user-times')
remove_btn.on_click(on_remove_click)
remove_box = widgets.HBox([remove_input, remove_btn])

# Tracking row
track_btn = widgets.Button(description='Start Analytics / Tracking', button_style='success', icon='play', layout=widgets.Layout(width='auto'))
track_btn.on_click(on_track_click)

# View Buttons row
view_users_btn = widgets.Button(description='View Users', button_style='info', icon='users')
view_att_btn = widgets.Button(description='View Logs', button_style='warning', icon='list')
view_chart_btn = widgets.Button(description='View Chart', button_style='info', icon='bar-chart')
clear_logs_btn = widgets.Button(description='Clear Logs', button_style='danger', icon='trash')

view_users_btn.on_click(on_view_users_click)
view_att_btn.on_click(on_view_attendance_click)
view_chart_btn.on_click(on_view_chart_click)
clear_logs_btn.on_click(on_clear_logs_click)

views_box = widgets.HBox([view_users_btn, view_att_btn, view_chart_btn, clear_logs_btn])

# Dashboard Layout
dashboard = widgets.VBox([
    widgets.HTML("<h2>Face Registration</h2>"),
    register_box,
    widgets.HTML("<hr>"),
    widgets.HTML("<h2>Remove User</h2>"),
    remove_box,
    widgets.HTML("<hr>"),
    widgets.HTML("<h2>Live Tracking</h2>"),
    track_btn,
    widgets.HTML("<hr>"),
    widgets.HTML("<h2>Data & Analytics</h2>"),
    views_box,
    widgets.HTML("<hr>"),
    output_area
])

display(dashboard)
```
