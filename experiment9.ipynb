```python
import pandas as pd
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import RandomUnderSampler

# 🔹 Load dataset
df = pd.read_csv("C:/Users/Anshika/.cache/kagglehub/datasets/mlg-ulb/creditcardfraud/versions/3/creditcard.csv")

# 🔹 Assume last column is target (change if needed)
target_col = df.columns[-1]

print("🔸 Target Column:", target_col)
print("\n🔸 Class Distribution Before:")
print(df[target_col].value_counts())

# =========================
# 🔤 ENCODING (if needed)
# =========================
df = pd.get_dummies(df, drop_first=True)

# Split features & target
X = df.drop(target_col, axis=1)
y = df[target_col]

# =========================
# ⚖️ 1. RANDOM OVERSAMPLING
# =========================
ros = RandomOverSampler(random_state=42)
X_ros, y_ros = ros.fit_resample(X, y)

print("\n🔸 After Random Oversampling:")
print(pd.Series(y_ros).value_counts())

# =========================
# ⚖️ 2. RANDOM UNDERSAMPLING
# =========================
rus = RandomUnderSampler(random_state=42)
X_rus, y_rus = rus.fit_resample(X, y)

print("\n🔸 After Random Undersampling:")
print(pd.Series(y_rus).value_counts())

# =========================
# ⚖️ 3. SMOTE (Best Method 🔥)
# =========================
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)

print("\n🔸 After SMOTE:")
print(pd.Series(y_smote).value_counts())

# =========================
# 💾 SAVE ONE RESULT (SMOTE)
# =========================
final_df = pd.concat([pd.DataFrame(X_smote), pd.DataFrame(y_smote, columns=[target_col])], axis=1)
final_df.to_csv('balanced_data.csv', index=False)

print("\n✅ Balanced dataset saved as 'balanced_data.csv'")
```
